```python
%%sql tsql << SELECT TOP (1000) 
      [City Key]
      ,[Total Including Tax]
into [WideWorldImportersDW].[dbo].[SalesTotals]
FROM [WideWorldImportersDW].[Fact].[Sale]
Order by [City Key]
```

```python
%sql select top 10 * from [WideWorldImportersDW].[dbo].[SalesTotals]
```

```python
%%sql tsql2 << SELECT Distinct [City Key] as city, sum([Total Including Tax]) as tot 
FROM [WideWorldImportersDW].[dbo].[SalesTotals]
Group By [City Key]
Order BY tot Desc
```
